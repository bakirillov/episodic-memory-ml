```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 14, 8
rcParams['font.size'] = 16
```

```python
df = pd.read_csv("english.csv", index_col=0)
wv = df.loc["crocodile"][0:-1].values
```

```python
df2 = pd.read_csv("english_1hot.csv", index_col=0)
ohot = df2.loc["crocodile"].values[0:-1]
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(2, 3)
histogram = fig.add_subplot(gs[0,0])
histogram.set_title("A. Noise, N(1,0.1)")
plt.hist(np.random.normal(size=300, loc=1, scale=0.1), bins=50)
ohot_plain = fig.add_subplot(gs[1,0])
ohot_plain.set_title("B. One-hot encoding")
plt.imshow(
    np.concatenate([(np.array([0]*26)).reshape(26, 1), ohot.reshape(26, 13)], 1).T
)
plt.xticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
plt.yticks(np.arange(len("crocodile")+3), " crocodile  ")
wv_plain = fig.add_subplot(gs[0,1:])
wv_plain.set_title("C. Plain word vector")
plt.imshow(wv.reshape(10, 30))
wv_aug = fig.add_subplot(gs[1,1:])
wv_aug.set_title("D. Augmented word vector")
plt.imshow((wv*np.random.normal(size=300, loc=1, scale=0.1)).reshape(10, 30))
plt.show()
```
