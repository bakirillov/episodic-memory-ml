```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 7, 4
```

```python
df = pd.read_csv("english.csv", index_col=0)
wv = df.loc["crocodile"][0:-1].values
```

```python
df2 = pd.read_csv("english_1hot.csv", index_col=0)
ohot = df2.loc["crocodile"].values[0:-1]
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(2, 3)
histogram = fig.add_subplot(gs[0,0])
histogram.set_title("A. Noize, N(1,0.1)")
plt.hist(np.random.normal(size=300, loc=1, scale=0.1), bins=50)
ohot_plain = fig.add_subplot(gs[1,0])
ohot_plain.set_title("B. One-hot encoding")
plt.imshow(ohot.reshape(26, 13))
#plt.xticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
#plt.yticks(np.arange(-2, len("crocodile")+2), "elidocorc")
wv_plain = fig.add_subplot(gs[0,1:])
wv_plain.set_title("C. Plain word vector")
plt.imshow(wv.reshape(10, 30))
wv_aug = fig.add_subplot(gs[1,1:])
wv_aug.set_title("D. Augmented word vector")
plt.imshow((wv*np.random.normal(size=300, loc=1, scale=0.1)).reshape(10, 30))
#ohot_plain = fig.add_subplot(gs[:,1])
#ohot_plain.set_title("D. One-hot encoding")
#plt.imshow(ohot.reshape(26, 13))
#plt.yticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
#plt.xticks(np.arange(len("crocodile")), "crocodile")
#ohot_aug = fig.add_subplot(gs[:,2])
#ohot_aug.set_title("E. Augmented one-hot")
#plt.imshow((ohot*np.random.normal(size=ohot.shape, loc=1, scale=0.1)).reshape(26, 13))
#plt.yticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
#plt.xticks(np.arange(len("crocodile")), "crocodile")
plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
histogram = fig.add_subplot(gs[0,0])
histogram.set_title("A. Noize, N(1,0.1)")
plt.hist(np.random.normal(size=300, loc=1, scale=0.1), bins=50)
wv_plain = fig.add_subplot(gs[1,0])
wv_plain.set_title("B. Plain word vector")
plt.imshow(wv.reshape(10, 30))
wv_aug = fig.add_subplot(gs[2,0])
wv_aug.set_title("C. Augmented word vector")
plt.imshow((wv*np.random.normal(size=300, loc=1, scale=0.1)).reshape(10, 30))
ohot_plain = fig.add_subplot(gs[:,1])
ohot_plain.set_title("D. One-hot encoding")
plt.imshow(ohot.reshape(26, 13))
plt.yticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
plt.xticks(np.arange(len("crocodile")), "crocodile")
ohot_aug = fig.add_subplot(gs[:,2])
ohot_aug.set_title("E. Augmented one-hot")
plt.imshow((ohot*np.random.normal(size=ohot.shape, loc=1, scale=0.1)).reshape(26, 13))
plt.yticks(np.arange(26), "abcdefghijklmnopqrstuvwxyz")
plt.xticks(np.arange(len("crocodile")), "crocodile")
plt.show()
```
