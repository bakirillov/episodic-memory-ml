```python
import json
from misc import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

# English

```python
with open("../models/PI/WV/english.csv.json", "r") as oh:
    english_results = json.load(oh)
```

```python
plt.scatter(english_results["train.aug.reals"], english_results["train.aug.preds"], color="blue")
plt.plot(english_results["train.aug.reals"], english_results["train.aug.reals"], color="red")
plt.scatter(english_results["test.reals"], english_results["test.preds"], color="green")
plt.scatter(english_results["train.reals"], english_results["train.preds"], color="purple")
plt.xlim((0,1))
plt.ylim((0,1))
plt.xlabel("Real AUROCs")
plt.ylabel("Predicted AUROCs")
plt.legend(
    loc="best",
    handles=[
        Patch(color="blue", label="Augmented pseudoexamples"),
        Patch(color="purple", label="Training set"),
        Patch(color="green", label="Test set"),
        Patch(color="red", label="Perfect prediction"),
    ]
)
plt.show()
```
